```python
def analyze_numbers(numbers):
    even_count=0
    odd_sum=0
    for num in numbers:
        if num%2==0:
            even_count+=1
        else:
            odd_sum+=num
    return even_count,odd_sum
nums=[1,2,45,67,7,9,3,4,6,8]
evens,odds=analyze_numbers(nums)
print(f"count of even numebrs:{nums}")
print(f"count of odd numbers:{odds}")
```
