```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length  # Saves the length
        self.width = width    # Saves the width

    # A method to calculate and print the area (length * width)
    def calculate_area(self):
        area = self.length * self.width
        print(f"--- Rectangle Area Calculation ---")
        print(f"Length: {self.length}")
        print(f"Width:  {self.width}")
        print(f"Area:   {area} sq units")
rect1 = Rectangle(10, 5)


# 3. Call the method to run the calculation
rect1.calculate_area()
```

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius  # Saves the radius

    def calculate_area(self):
        pi = 3.14  
        area = pi * (self.radius ** 2)
        
        print(f"--- Circle Area Calculation ---")
        print(f"Radius: {self.radius}")
        print(f"Area:   {area} sq units")
circle1 = Circle(7)
circle1.calculate_area()
```

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance  # Storing the starting balance

    def deposit(self, amount):
        self.balance += amount  # Shortcut for: self.balance = self.balance + amount
        print(f"💰 Deposited: ₹{amount}")
        print(f"Current Balance: ₹{self.balance}")
        print("-" * 30)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount  # Shortcut for: self.balance = self.balance - amount
            print(f"💸 Withdrew: ₹{amount}")
            print(f"Current Balance: ₹{self.balance}")
        else:
            print(f"❌ Transaction Failed: Insufficient funds for withdrawal of ₹{amount}!")
            print(f"Current Balance remains: ₹{self.balance}")

account1 = BankAccount("Rahul", 5000)

print(f"Initial Account Holder: {account1.account_holder}")
print(f"Starting Balance: ₹{account1.balance}")

account1.deposit(1500)   # Deposits 1500
account1.withdraw(2000)  # Withdraws 2000
account1.withdraw(6000)  # Tries to withdraw more than available
```

```python
class Animal:
    def sound(self):
        print("This animal makes a generic sound.")


class Dog(Animal):  
    def sound(self):
        print("Bark")


generic_animal = Animal()
my_dog = Dog()

print("Parent Class Sound:")
generic_animal.sound()  
print("-" * 25)

# Call the sound method on the dog
print("Child Class (Overridden) Sound:")
my_dog.sound()  # Outputs the specific overridden "Bark" statement
```

```python
# 1. Define the Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name  # Parent handles name
        self.age = age    # Parent handles age


# 2. Define the Child Class (Inherits from Person)
class Student(Person):
    def __init__(self, name, age, roll_number):
        # super() links to Person and runs its constructor
        super().__init__(name, age)
        self.roll_number = roll_number  # Student handles roll number

    # Method to display all details combined
    def display_student_details(self):
        print(f"--- Student Record ---")
        print(f"Name:        {self.name}")
        print(f"Age:         {self.age}")
        print(f"Roll Number: {self.roll_number}")
        print("-" * 22)


# 3. Create an object of the Student class
student_obj = Student("Rahul Sharma", 20, "CS-2026-05")


# 4. Display the details using the child object
student_obj.display_student_details()
```

```python
# 1. Define the Calculator blueprint
class Calculator:
    def add(self, num1, num2):
        result = num1 + num2
        print(f"Addition:       {num1} + {num2} = {result}")

    def subtract(self, num1, num2):
        result = num1 - num2
        print(f"Subtraction:    {num1} - {num2} = {result}")

    def multiply(self, num1, num2):
        result = num1 * num2
        print(f"Multiplication: {num1} * {num2} = {result}")

    def divide(self, num1, num2):
        if num2 != 0:
            result = num1 / num2
            print(f"Division:       {num1} / {num2} = {result}")
        else:
            print("Division:       Error! Cannot divide by zero.")


# 2. Take two numbers as input from the user
print("--- Python Calculator App ---")
n1 = float(input("Enter first number: "))
n2 = float(input("Enter second number: "))
print("=" * 30)

# 3. Create a calculator object
my_calc = Calculator()

# 4. Perform all operations using the object
my_calc.add(n1, n2)
my_calc.subtract(n1, n2)
my_calc.multiply(n1, n2)
my_calc.divide(n1, n2)
print("-" * 30)
```

```python
# 1. Define the LibraryBook blueprint
class LibraryBook:
    def __init__(self, book_name, author, price):
        self.book_name = book_name  # Saves the book name
        self.author = author        # Saves the author name
        self.price = price          # Saves the book price

    # Method to print all details
    def display_book_info(self):
        print(f"--- Book Details ---")
        print(f"Title:  {self.book_name}")
        print(f"Author: {self.author}")
        print(f"Price:  {self.price}")
        print("-" * 25)


# 2. Create three separate book objects
book1 = LibraryBook("To Kill a Mockingbird", "Harper Lee", "₹399")
book2 = LibraryBook("1984", "George Orwell", "₹299")
book3 = LibraryBook("The Hobbit", "J.R.R. Tolkien", "₹450")


# 3. Display information for all three books
book1.display_book_info()
book2.display_book_info()
book3.display_book_info()
```
